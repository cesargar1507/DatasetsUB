```r
# Ejercicio 1: Creación de la función roll para simular un dado de 20 caras
roll <- function() {
  return(sample(1:20, 1, replace=TRUE))
}
```

```r
# Ejercicio 2: Simulación de 50,000 tiradas de un dado de 20 caras
results <- rep(0, 20)
num_rolls <- 50000

for (i in 1:num_rolls) {
  result <- roll()
  results[result] <- results[result] + 1
}

barplot(results / num_rolls * 100, main="Resultados de 50,000 tiradas de un dado de 20 caras", xlab="Resultado del dado", ylab="Porcentaje (%)", names.arg=1:20)
```

```r
# Escenarios 1 y 2
scenario1 <- function() {
  score <- sum(roll(), 5000)
  return(score > 25000)
}

scenario2 <- function() {
  score <- sum(replicate(5000, sum(roll(), 5)))
  return(score > 125000)
}
```

```r
# Ejercicio 3: Simulación de 5000 tiradas de 1 dado de 20 caras
results_scenario1 <- rep(0, 20)

for (i in 1:5000) {
  result <- roll()
  results_scenario1[result] <- results_scenario1[result] + 1
}

barplot(results_scenario1 / 5000, main="Resultados de 5000 tiradas de 1 dado de 20 caras", xlab="Resultado del dado", ylab="Frecuencia", names.arg=1:20)

# Ejercicio 4: Simulación de 5000 tiradas de 5 dados de 20 caras y mostrar el promedio
results_scenario2 <- rep(0, 20)  # Inicializamos un vector para almacenar los resultados

for (i in 1:5000) {
  score <- mean(replicate(5, roll()))  # Calculamos el promedio de los resultados de los 5 dados
  results_scenario2[round(score)] <- results_scenario2[round(score)] + 1  # Aumentamos la frecuencia del promedio obtenido
}

# Generamos el gráfico de barras
barplot(results_scenario2 / 5000, main="Resultados de 5000 tiradas de 5 dados de 20 caras (Promedio)",
        xlab="Promedio de los resultados de los 5 dados", ylab="Frecuencia", names.arg=1:20)
```

```r
percentages_dice1 <- round(results_scenario1 / sum(results_scenario1) * 100, 2)
percentages_dice5 <- round(results_scenario2 / sum(results_scenario2) * 100, 2)
print("Escenario 1")
print(data.frame(Resultado = 1:20, Porcentaje = percentages_dice1))
print("Escenario 2")
print(data.frame(Resultado = 1:20, Porcentaje = percentages_dice5))
```

Es más probable superar la tirada de salvación al arrojar 5 dados, los datos parecen seguir una distribución normal, hay una tendencia a que el promedio en las tiradas sea de 11

```r
loaded_dice <- function() {
  probabilities <- c(2/25, 2/25, 2/25, 2/25, 2/25, rep(1/25, 15))
  return(sample(1:20, 1, prob=probabilities))
}
```

```r
# Ejercicio 7: Simulación de 5000 tiradas de 1 dado trucado de 20 caras
results_loaded_dice1 <- rep(0, 20)

for (i in 1:5000) {
  result <- loaded_dice()
  results_loaded_dice1[result] <- results_loaded_dice1[result] + 1
}

barplot(results_loaded_dice1 / 5000, main="Resultados de 5000 tiradas de 1 dado trucado de 20 caras",
        xlab="Resultado del dado", ylab="Frecuencia", names.arg=1:20)

# Ejercicio 8: Simulación de 5000 tiradas de 5 dados trucados de 20 caras y mostrar el promedio
results_loaded_dice5 <- rep(0, 20)

for (i in 1:5000) {
  score <- mean(replicate(5, loaded_dice()))  # Calculamos el promedio de los resultados de los 5 dados
  results_loaded_dice5[round(score)] <- results_loaded_dice5[round(score)] + 1  # Aumentamos la frecuencia del promedio obtenido
}

# Generamos el gráfico de barras
barplot(results_loaded_dice5 / 5000, main="Resultados de 5000 tiradas de 5 dados trucados de 20 caras (Promedio)",
        xlab="Promedio de los resultados de los 5 dados", ylab="Frecuencia", names.arg=1:20)
```

```r
percentages_loaded_dice1 <- round(results_loaded_dice1 / sum(results_loaded_dice1) * 100, 2)
percentages_loaded_dice5 <- round(results_loaded_dice5 / sum(results_loaded_dice5) * 100, 2)
print("Escenario 1")
print(data.frame(Resultado = 1:20, Porcentaje = percentages_loaded_dice1))
print("Escenario 2")
print(data.frame(Resultado = 1:20, Porcentaje = percentages_loaded_dice5))
```

Es más probable ganar al jefe final haciendo tiradas de 1 solo dado, al observar las frecuencias a la derecha de 15 se observa una mayor cantidad de datos en el primer escenario, para el segundo escenario, los datos parecen seguir una distribución normal, tienden en promedio a 8, la frecuencia de tiradas que superan los 15 es poca (no es superior al 1%)
